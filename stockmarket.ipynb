```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('stock_raw_data.csv')
df.head()
```

```python
# correlation matrix
corr = df.corr()
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# distribusi adj_high and adj_low
sns.displot(df, x="adj_high", kde=True)
sns.displot(df, x="adj_low", kde=True)
```

```python
# bivariate distribution adj_close and volume
sns.displot(df, x="adj_close", y="volume", kind="kde")
```
